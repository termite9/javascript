### 2장 데이터 형과 연산자 (p.40)
- 변수이름짓기
- 데이터 형 - 숫자,문자열,불,Null,Undefined,Bigint,객체,심볼 여덟가지가 있다.
- 연산자    - 숫자와 문자열 등을 계산하기 위한 산술연산자, 문자열 연결연산자, 할당연산자,비교연산자, 논리 연산자에 대해 학습한다.
- 기초프로그램 작성법

#### 2.1 변수란?
- 데이터가 컴퓨터 메모리에 저장되는 주소를 말한다.

> 2.1.1 변수 선언
- var,let,const 세개의 키워드가 있다. var는 2015년 이전에 사용되었던 변수로  이 책의 모든 예제들은 let과 const를 사용한다.
- 

- * let과 달리 const는 constant(상수)의 약어로 값이 변하지 않는 수를 의미합니다. 따라서 const변수는 그 값이 고정된 변수에 사용합니다. 재할당하면 오류가 발생합니다.

> 2.1.1 변수 이름 짓기  변수이름을 식별자라고 한다.
- 영어 알파벳,숫자,밑줄(_),$를 사용합니다.
- 숫자로 시작하면 안된다.
- 알파벳 대소문자를 구별한다.
- 길게 지을때는 _이나 캐멀케이스가 사용된다

#### 2.2 데이터 형 

> 2.2.1 숫자
- 정수(integer)와 부동소수점 숫자(floating point number)가 사용된다.
- 

- 다른 컴퓨터 언어와 달리 정수와 부동소수점 숫자 모두 64비트 부동소수점으로 저장된다(이게 무슨 말인지는 모른)

> 2.2.2 Bigint
- 자바스크립트에서 정수는 15자리까지만 나타낼 수 있다. 그이상의 정수를 사용하려면 Bigint형을 사용해야 한다.

> 2.2.3 문자열
- 큰따옴표 ""나 작은따옴표''로 감싸야한다
- 문자열의 length프로퍼티를 이용하면 문자열의 길이를 구할 수 있다
- 공백도 하나의 문자라는 것을 꼭 기억하기 바란다.
- 따옴표안에 따옴표를 사용하면 오류가 발생한다. 이럴때는 "내 이름은 \"홍길동\" 입니다" 와같이 해주면 된다.
- 위의 예에서 역슬래쉬(\)가 사용되는 문자를 이스케이프문자라고 한다. 많이 사용되는 것중에 \n 은 줄바꿈을 나타낸다
- 특수코드를 사용하려면 복붙을 하거나 해당코드를 타이핑하면 된다.
