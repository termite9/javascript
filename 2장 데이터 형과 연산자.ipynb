> 2.2.4 템플릿 문자열
- 문자열 안에 직접 변수 값을 넣을 수 있다.

```python
 <script>
        let name = "홍길동"
        let age = 20

        let text1 = "이름: " + name + ",나이:" + age;
        let text2 = `이름: ${name}, 나이: ${age}`;
        document.write(text1);
        document.write(text2);
    </script>
```

- 템플릿 문자열에슨 따옴표 대신 백틱(``)를 사용하는 것 이 특징이다. 백틱안에 있는 ${}는 문자열로 변환되어 백틱 내의 다른 문자열과 합쳐진다.

> 불 bool
- 불데이터 형에는 true와 false만 존재합니다.

```python
<script>
let x = 5 > 3;
let y = 5 < 3;

document.write(x + "<br>");
document.write(y);
</script>
```

> 2.2.6 Undefined와 Null
-  변수가 값을 가지고 있지 않으면 그 값은 undefined가 됩니다.

>Undefined와 달리 Null데이터 형은 빈문자열을 의미합니다.

```python
<script>
let x = "";
document.write("값:" + x + "<br>");
document.write("데이터형: " + typeof x);
</script>
```


- x는 빈 문자열이기 때문에 화면에 값을 출력하지 않습니다. 다만 빈 문자열을 의미하기 때문에 일종의 문자열(string)이라고 볼 수 있습니다.

#### 2.3 연산자 

> 2.3.1 산술연산자
- %(나머지계산), ++(1씩증가), --(1씩 감소)

> 2.3.2 문자열 연결 연산자
- 산술연산자에서 사용되는 덧셈연산자  + 가 그대로 사용된다
- "안녕" + "하세요"

- 문자열 "홍길동"과 정수형 나이 30을 더하면 문자열이된다. "홍길동"+30 = "홍길동30"
- 이것은 신기한게 아니고 자바스크립트의 맹한 점이다.ㅋ

> 2.3.3 할당연산자
- 데이터나 변수 값을 변수에 저장, 즉 메모리 공간에 할당하는 역할을 수행합니다.
    - =  (x = 2) : 2를 변수 x에 할당
    - += (x+=2)  : x=x+2
    - -= (x-=2)  : x=x-2
    - *= (x*=2)  : x=x*2
    - /= (x/2=2) : x=x/2
    - %= (x%=2)  : x= x%2

```python
<script>
//할당연산자 문자열에 적용된예
let name = "홍길동";
let tel = '010-9171-9616';
let age = 50;

let text = "";
text += "<table border = '1'>";
text += "<tr><th>이름</th><th>전화번호</th><th>나이</th></tr>";
text += "<tr><td>" + name + "</td><td>" + tel + "</td><td>" + age + "</td></tr>";

document.write(text);
</script>
```

#### 2.4 프로그래밍 맛보기 
- 변수,연산자,데이터형을 활용하여 간단한 자바스크립트 프로그램을 작성해본다

> 2.4.1 원의 넒이 구하기

```python
<script>
    let r = 8, area; //반지름으로 8을 넣고 넒이를 나타내는 변ㄴ수 area를 선언합니다.
    const pi = 3.14;
    area = r*r*pi;

    document.write("반지름:" + r + "<br>");
    document.write("원의넒이:" + area);
</script>
```

> 2.4.2 거스름돈 구하기

```python
<script>
        let pay = 10000; // 지불금액
        let price = 1500; //물건가격
        let num = 5; //구매개수
        let change; //거스름돈

        change = pay - price * num;

        document.write(change);
    </script>
```

> 2.4.3 섭씨/화씨 변환하기
- 섭씨를 화씨로 변환하는 공식 : 화씨 = 섭씨 * 9/5 + 32

```python
<script>
        let c = 30; //섭씨온도
        let f; //화씨온도

        f = c * 9 / 5 + 32;
        document.write("섭씨:" + c + "는 화씨:" + f);
    </script>
```
