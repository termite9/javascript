#### 1.1 자바스크립트와 개발도구
- 자바스크립트는 웹 페이지에서 마우스로 메뉴 버튼을 클릭했을 때 매뉴가 펼쳐지거나 키보드를 클릭했을대 웹페이지 요소의 움직임을 제어하는등 웹 페이지에서 일어나는 모든 동작을 제어한다.

> 1.1.1 웹과 자바스크립트
- 웹 브라우저에서 사용할 수있는 유일한 컴퓨터 언어이다.
- 자바 스크립트로 작성한 프로그램을 스크립트 Script라고 합니다.
- 기계어로 바꾸어 주는 컴파일 과정을 거치지 않는다.
- 자바 스크립트를 기반으로 한 node.js는 서버측 에서도 자바스크립트를 그대로 사용하며 모든 기능을 구현 할 수있게 해준다.

> 1.1.2 개발도구
- 비주얼스튜디오코드 사용한다.
